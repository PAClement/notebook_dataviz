```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Matplot

```python
# Plot a point

x = 4

y = 10


plt.scatter(x,y)
```

```python
# Multiple points, in red

x= [1, 4, 12, 8]
y = [9, 12, 3, 4.9]

plt.scatter(x,y, c= 'black')
```

```python
# Plot a line, you need two points to draw a line

x = [1, 3]
y = [2, 6]

plt.plot(x,y, color='brown')
```

```python
# Two lines with a legend

x_1 = [1, 3]
y_1 = [2, 6]

x_2 = [3, 8]
y_2 = [1.5, 9]

plt.plot(x_1,y_1, color='brown', label='first')

plt.plot(x_2,y_2, color='green', label='second')

# Change the length of y and x axis
plt.xlim(0,10)
plt.ylim(0, 10)

plt.legend() # legend is used to show labels and differentiate between different plots
```

# Seaborn

```python
df = sns.load_dataset('tips')

df.head()
```

### Barplot (categorical + Numerical)

- SQL : Count(*), Group BY (day)

```python
# Count data points using bars (diagramme  bandes), and specify the order

sns.countplot(data = df, x='day', order=['Thur', 'Fri', 'Sun', 'Sat'])

plt .title("Number of Tables by day")
```

#### What do you understand from the plot

```python
# Use hue to breakdown by gender

sns.countplot(data = df, x='day', hue='sex')
```

```python
# Show how to get code, step by step

df.sex.value_counts().index
```

```python
df.sex.value_counts()
```

### Pie Chart to show proportions

***Globally, males are overrepresented***

```python
plt.pie(x = df.sex.value_counts(), labels=df.sex.value_counts().index, autopct='%1.1f%%'); # autopct to write percentages
```

```python
df.time.value_counts()
```

```python
# Let's use subplots to plot three different graphs beside each others

plt.figure(figsize = (14,5))

# First plot, 1 out of 3
plt.subplot(1,3,1) # subplot(num_rows, num_columns, num_plot)
plt.pie(df.time.value_counts(), labels=df.time.value_counts().index)
plt.title('Time Percentage')


# Second plot
plt.subplot(1,3,2)
sns.histplot(x='total_bill', data= df)
plt.title('Distribution of Total Bill')

# Third plot
plt.subplot(1,3,3)
plt.scatter(df['total_bill'], df['tip'])
plt.title('Correlation Between Bill and Tip')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
```

```python
# Histogram to show distribution of data

sns.histplot(df.tip, kde=True);
plt.xticks(np.arange(10));   # specify numbers to show on the X axis
plt.yticks(np.arange(0,60,5)); # same as xticks but for Y axis
```

```python
# Show distribution of tip by day using Box graph

# Note : catplot is used to show relationship between a numerical and one or more categorical variables

sns.catplot(data=df, x='day', y='tip', kind='box')
```
